# Spliting, Joining, Formatteed String

```python
prompt = "What is Python ?"
tokens = prompt.split()
print(tokens)
```

```python
prompt = "What-is-Python-?"
tokens = prompt.split("-")
print(tokens)
```

```python
# now we will convert these tokens into a string
sentence = "-".join(tokens)
print(sentence)
```

# List

```python
# list is mutable but string is immutable
# means 1st bar value assign krar poreo edit kora jay list a

numbers = [10,28,34,65]
numbers[2] = 5
print(numbers)
```

```python
# adding a item at last
numbers.append(70)
print(numbers)
```

```python
# adding number at specific position

numbers.insert(0, 100) # at 0th position
print(numbers)
```

```python
# deleting from last
numbers.pop()
print(numbers)
```

```python
# remove only first number:

numbers.remove(100)
print(numbers)
```

```python
numbers.reverse()
print(numbers)
```

```python
numbers.sort()
print(numbers)
```

```python
numbers.sort(reverse=True)
print(numbers)
```

# Queue

```python
# 1st e je input hoy, take niye sobar age kaj hoy.
# notun element last e insert hoy
# 1st jon ke remove korle 2nd jon 1st joner jaygay ashe, similarly 3rd jon 2nd joner jaygay ashe
```

```python
queue = []

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

#removing 1st one
print(queue.pop(0))

print(queue)
```

# 3.7 List Comprehension

```python
even = []
for i in range(2,98,1):
    if(i%2==0):
        even.append(i)

print(even)
```

```python
# we can do this using list comprehension

random_list = [i for i in range(11) if i%2==0]
print(random_list)
```

# 3.8 List Comprehension with List of String

```python
fruits = ['apple', 'orange', 'mango']

capitalize_fruits = []
for fruit in fruits:
    capitalize_fruits.append(fruit.capitalize())

print(capitalize_fruits)
```

```python
fruits = ['apple', 'orange', 'mango']

upper_fruits = [i.upper() for i in fruits]

print(upper_fruits)
```
