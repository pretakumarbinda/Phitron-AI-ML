# Iterator

```python

arr = np.array([12,32,42,13,453,12])
arr_iter = iter(arr)

print(next(arr_iter))
print(next(arr_iter))
print(next(arr_iter))
print(next(arr_iter))
```

# Generator



```python
np.arange(20)
```

```python


def data_loader(batch_size, data):
    for i in range(0, len(data), batch_size):
        yield data[i: i+batch_size]      # yield -> return chunk, not whole data

data = np.arange(20)

loader = data_loader(5, data)

print(next(loader))
```

```python
print(next(loader))
```
