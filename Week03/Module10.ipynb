# transforming to df from inbuilt dates

```python
my_list = [['Alice', 24], ['Bob', 27], ['Charlie', 22]]

list_df = pd.DataFrame(my_list, columns=["Name","Age"], index=['a', 'b', 'c'])
list_df
```

```python
my_dict = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [24, 27, 22]}

dict_df = pd.DataFrame(my_dict) # automatically column name will be the key of the dictionary
print(dict_df)
```

# Accessing values of a dataframe

```python
# accessing the column of the dataframe
print(df['StudentID'])
```

```python
# accessing the row of the dataframe
# loc and iloc are used to access the row of the dataframe

#loc[row(index), column(features)]
# df.loc[ind_start:ind_end, col_start:col_end]

df.loc[0] # this will return the first row of the dataframe
```

```python
df.loc[0:3] # this will return the first 3 rows of the dataframe
```

```python
df.loc[0:2, 'StudentID'] # this will return the first 3 rows of the StudentID column
```

```python
# single column access
df.loc[:, "Python Marks"]
```

```python
# multiple column access
df.loc[:, ["Python Marks", "StudentID"]]
```

# Changing values & Iteration

```python
df.head()
```

```python
df = pd.read_csv('student_data.csv')
df.drop(0, inplace=True) # this will drop the row with index 0
```

```python
df
```

```python
# drop column
df.drop('Instructor', axis = 1, inplace=True)
df
```

```python
# value change

df.loc[1, 'Python Marks'] = 95
df
```

```python
df.loc[1:3, 'Python Marks']+=4
df.head()
```

```python
df.loc[:, 'Python Marks']-=5
df
```

# sort

```python
df.sort_values(by='Python Marks', ascending=False, inplace=True)
df
```

```python
copy = df.sort_values(['Python Marks', 'Algorithm Marks'], ascending=[False, True])  # this will sort the dataframe by Python Marks in descending order and if there are multiple rows with the same Python Marks, then it will sort those rows by algorithm Marks in ascending order
```

```python
copy
```

# Filtering data based on condition

```python
df
```

```python
not_started =df.loc[df['CompletionStatus']=='Not Started']
not_started
```

```python
completed = df.loc[df['CompletionStatus']=='Completed']

completed
```
