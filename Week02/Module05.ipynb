```python
class Phone:
    category = "Electronics" #class attribute
    
    #constructor
    def __init__(self,model,battery,camera, battery_percentage=100):
        self.model= model
        self.camera = camera
        self.battery = battery_percentage
        
    # methods
    def charge(self,hour):
        self.battery += hour
    
    def capture(self,total_photo):
        self.battery -= total_photo
        print("photo captured")
```

#### Inheritance

```python
# Phone class -> parent class
# Smartphone -> child class

class Smartphone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery,camera)    # parent class er init ke call kora hoise. ejonne ekhane processor nai
        self.processor = processor
    
    def charge(self,hour):
        print("fast charging in process")
        super().charge(hour)
```

```python
pro = Smartphone("X", 5000, 40, "snapdragon")
print(pro.model)
pro.charge(1)
```

### multiple inheritance

```python
class Phone:
    category = "Electronics" #class attribute
    
    #constructor
    def __init__(self,model,battery,camera, battery_percentage=100):
        self.model= model
        self.camera = camera
        self.battery = battery_percentage
        
    # methods
    def charge(self,hour):
        self.battery += hour
    
    def capture(self,total_photo):
        self.battery -= total_photo
        print("photo captured")
        
    
class Cooling_Method:
    def __init__(self,cooling_method):
        self.cooling_method = cooling_method
    
    # method
    def cooling_on(self):
        print(f"System is being cooled by {self.cooling_method}")
```

```python
# Phone class -> parent class
# Smartphone -> child class

class Smartphone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery,camera)    # parent class er init ke call kora hoise. ejonne ekhane processor nai
        self.processor = processor
    
    def charge(self,hour):
        print("fast charging in process")
        super().charge(hour)
```

```python
# here in this child class we will use both cooling_method class and smartphone class:

class Smart_cooling_phone(Smartphone,Cooling_Method):
    def __init__(self, model, battery, camera, processor,cooling_method):
        Smartphone.__init__(self, model, battery,camera, processor)  # multi class inheritance er somoy upper() use na kore direct class er name use hobe
        Cooling_Method.__init__(self, cooling_method)
        
```

```python
pro_cooling = Smart_cooling_phone("Y", 4000, 50,"Snapdragon","Nitrozen")
print(pro_cooling.processor)        # from SmartPhone class
print(pro_cooling.battery)          # from Phone class
print(pro_cooling.cooling_method)   # from Cooling_Method class
print(pro_cooling.charge(2))        # from smartphone er modified charge jeta phone class theke smartphone e inherited
print(pro_cooling.cooling_on())     # from cooling_Method class
```

# Polymorphism

```python
class Camera:
    def __init__(self,name):
        self.name = name
    
    # captured method:
    def captured(self):
        print("A photo is captured")
```

```python
class smart(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
    # capture method overriding
    def captured(self):
        print("Phonto captured by a phone")

class DSLR(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
        # capture method overriding
    def captured(self):
        print("Phonto captured by a DSLR")

class Drone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resulation = resulation
    
    # capture method overriding
    def captured(self):
        print("Phonto captured by a drone")

phone = smart("Nothing 3A", 50)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 70)

phone.captured()
dslr.captured()
drone.captured()
```

# Encapsulation

Encapsulation মানে হলো:
👉 data (attribute) + method একসাথে রাখা এবং data-কে direct access থেকে protect করা

সহজভাবে:
👉 data hide করা + controlled access দেওয়া

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.imei = imei
    def charge(self):
        print("phone is charging")

Iphone = Mobile("phone", "17", "efwe33f")

print(Iphone.name)
print(Iphone.model)
Iphone.name = "Preta"       # change kora jay bahir theke
print(Iphone.name)
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei      # private now
    def charge(self):
        print("phone is charging")

Iphone = Mobile("phone", "17", "efwe33f")

print(Iphone.name)
print(Iphone.model)
Iphone.name = "Preta"       # change kora jay bahir theke
print(Iphone.name)

print(Iphone.__imei)        # cann't be accessed, and cann't be changed
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei      # private now
    def charge(self):
        print("phone is charging")
    def imei_get(self):         # for accessing imei
        return self.__imei

Iphone = Mobile("phone", "17", "efwe33f")

print(Iphone.name)
print(Iphone.model)
Iphone.name = "Preta"       # change kora jay bahir theke
print(Iphone.name) 

print(Iphone.imei_get())        # now can be accessed
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.__name = name
        self.__model = model
        self.__imei = imei      # private now

    def name_getter(self):
        return self.__name
    def name_setter(self, name):
        self.__name = name

phone = Mobile("Nothing 3A", "17", "efwe33f")
print(phone.name_getter())
phone.name_setter("Iphone 18 pro")      # thus, protected things can be changeed
print(phone.name_getter())



```
