# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2,3,4,1,5,7,3]
days = [ 1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
```

```python
hours = [2,3,4,1,5,7,3]
days = [ 1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
plt.xlabel('Day')
plt.ylabel('Hours')
plt.title('Study Hours per Day in a week')
```

```python
hours_1 = [2,3,4,1,5,7,3]
hours_2 = [1,2,3,4,5,6,7]
days = [ 1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours_1, label='Student 1')
plt.plot(days, hours_2, label='Student 2')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.title('Study Hours per Day in a week')
plt.legend()
```

# 2D line plot from pandas

```python
df = pd.read_csv('enrollment_data.csv')
```

```python
df.head()
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'],df['Digital Marketing'], label='Digital Marketing')
plt.xlabel('Year')
plt.ylabel('Number of students')
plt.title('Enrollment Data')
plt.legend(loc='upper left')
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming',color='red',linewidth=4)
plt.plot(df['Year'],df['Digital Marketing'], label='Digital Marketing',color='blue')
plt.xlabel('Year')
plt.ylabel('Number of students')
plt.title('Enrollment Data')
plt.legend(loc='upper left')
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming',color='red',linewidth=2,marker='o',markersize=8)
plt.plot(df['Year'],df['Digital Marketing'], label='Digital Marketing',color='blue',marker='s',markersize=8)
plt.xlabel('Year')
plt.ylabel('Number of students')
plt.title('Enrollment Data')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# scatter plot

to find correlation between features

- positive correlation
- negative correlation


```python
students = pd.read_csv('student_IQdata.csv')
students.head()
```

```python
plt.scatter(students['Study_Hour'], students['IQ_Score'])
plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Study Hours vs IQ Score')
plt.show()
```

```python
plt.scatter(students['Chilling_Hours'], students['IQ_Score'], c='blue', marker='x')
plt.xlabel('Chilling Hours')
plt.ylabel('IQ Score')
plt.title('Chilling Hours vs IQ Score')
plt.show()
```

# Histogram
- how the data is distributed over a factor
- frequency
- Outliers

distribution:
- left-skewed distribution
- right skewed distribution
- symmetrical distribution (Bell curve)

```python
plt.hist(students['IQ_Score'], color='green')
plt.xlabel('IQ Score')
plt.ylabel('Frequency')
plt.title('Distribution of IQ Scores')
plt.show()
```

```python
plt.hist(students['Study_Hour'], color='green', edgecolor='black')
plt.xlabel('Study Hours')
plt.ylabel('Frequency')
plt.title('Distribution of Study Hours')
plt.show()
```

```python
plt.hist(students['Study_Hour'], color='green', edgecolor='black', bins=[0,2,4,6,8,10,12,14,16])
plt.xlabel('Study Hours')
plt.ylabel('Frequency')
plt.title('Distribution of Study Hours')
plt.show()
```

# Barchart

- x axis - categorical value


```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# gender wise

gender_group = df.groupby('Gender').size()
gender_group
```

```python
gender_group.index
```

```python
gender_group.values
```

```python
plt.bar(gender_group.index, gender_group.values,color='skyblue')    # category in x axis and frequency in y axis
plt.xlabel('Gender')
plt.ylabel('Frequency') 
plt.title('Gender Wise Distribution')
plt.show()
```

# Pie chart

```python
df = pd.read_csv('student_data_modified.csv')
```

```python
df
```

```python
status = df.groupby('CompletionStatus').size()
plt.pie(status.values, labels=status.index, autopct='%1.1f%%')
plt.title('Completion Status Distribution')
plt.show()
```

```python
status = df.groupby('CompletionStatus').size()
plt.pie(status.values, labels=status.index, autopct='%1.1f%%',explode=[0.1,0,0])
plt.title('Completion Status Distribution')
plt.show()
```

| Graph Type       | কী বোঝায়                                        | কখন ব্যবহার করবে                | Example use case      |
| ---------------- | ----------------------------------------------- | ------------------------------- | --------------------- |
| **Histogram**    | ডেটার distribution (কতবার কোন range এ মান পড়ছে) | Continuous data spread দেখতে    | Age distribution      |
| **Scatter Plot** | দুইটা ভেরিয়েবলের সম্পর্ক (relationship)         | Correlation বা pattern খুঁজতে   | Height vs Weight      |
| **Bar Chart**    | category অনুযায়ী মান compare                    | Discrete category comparison    | Product sales         |
| **Line Plot**    | সময়ের সাথে change (trend)                       | Time series data                | Stock price over time |
| **Pie Chart**    | percentage / proportion                         | Whole-এর মধ্যে অংশ দেখাতে       | Market share          |
| **Box Plot**     | distribution + outliers                         | Median, quartile, outlier দেখতে | Salary spread         |
| **Heatmap**      | intensity / correlation                         | Matrix data visualization       | Correlation matrix    |
| **Area Plot**    | cumulative change                               | Trend + volume দেখাতে           | Population growth     |
