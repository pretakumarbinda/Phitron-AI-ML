# Iterators

In Python, an iterator is an object that lets you go through (iterate over) a collection of data one element at a time. List has index values, but set cann't.

```python
s = [10, 20,12,321]

iterable_s = iter(s)
print(next(iterable_s))
print(next(iterable_s))
print(next(iterable_s))
print(next(iterable_s))
```

thus we can iterate them one by one in a list. but for a dictionary it is not serially iterated.

# Generator

Generator হলো এমন একটা বিশেষ ধরনের iterator, যা একসাথে সব ডাটা মেমোরিতে রাখে না—বরং একটা একটা করে value তৈরি করে দেয় (lazy evaluation)।

👉 এটা মূলত function দিয়েই তৈরি করা হয়, কিন্তু return এর জায়গায় yield ব্যবহার করা হয়।

```python
list1 = [x for x in range(100)]  # dataset

def data_loader(chunk_size, dataset):
    for i in range(0, len(dataset), chunk_size):
        yield dataset[i:i+chunk_size]

chunked_dataset = data_loader(5,list1)

print(type(chunked_dataset))
print(next(chunked_dataset))
print(next(chunked_dataset))
print(next(chunked_dataset))
```

# Some functions

```python
# lambda function
import pandas as pd
df = pd.DataFrame({'Temperature(F)':[32, 56, 232, 434]})
print(df)
```

Lambda function হলো Python-এর একটা ছোট, এক লাইনের anonymous function (নাম ছাড়া function)। <br>
<p>
syntax: <br>
        lambda arguments: expression </p>

```python
add = lambda a, b: a + b
# lambda arguments: expression

print(add(3, 5))  # 8
```

```python
df['Temperature(C)'] = df["Temperature(F)"].apply(lambda x: (x-32)*(5/9))
df
```

```python
df2 = pd.DataFrame({'A': [1,2,3,4,5], 'B':[6,7,8,9,10]})
df2
```

```python
df2['root'] = df2['A'].apply(lambda x: x**(1/2))
df2
```

```python
df3 = pd.DataFrame({'TK': [100,600,1000,344]})

df3['expensive'] = df3['TK'].apply(lambda x: 1 if x>500 else 0)
print(df3)
```
