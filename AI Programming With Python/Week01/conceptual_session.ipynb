# Lambda, map, filter, reduce func

```python
def greet():
    print("hello")
    
greet()
```

```python
greet = lambda : print("Hello")
greet()
```

```python
# with argument
def greet(name):
    print("hello ", name)
    
greet("preta")
```

```python
greet = lambda name : print("Hello", name)
greet("preta")
```

```python
greet = lambda name, age: print(f"Hi! I am {name} and I am {age} years old")

greet("preta", "22")
```

# Map
map(function, iterable, ...)

Examples of Iterables

- Lists: [1, 2, 3]
- Tuples: (1, 2, 3)
- Strings: "hello"
- Dictionaries: {'a': 1, 'b': 2}
- Sets: {1, 2, 3}
- Files: open('file.txt')

```python
arr = [1,2,3,4,5,6]

n = list(map(lambda x: x*10, arr))
print(n)
```

```python
scores = [85, 92, 78, 65, 45]

def grade(x):
    if x >=80 : return f"Score is {x} and A+"
    if x >=70 : return f"Score is {x} and A"
    if x >=60 : return f"Score is {x} and A-"
    else : return f"Score is {x} and F"
    
ans = list(map(grade, scores))
print(ans)
```

# filter
filter(function, iterable)

```python

scores = [85, 92, 78, 65, 45]

ans = list(filter(lambda x: x>80,scores))
print(ans)
```

filter returns the true value after seeing the condition.
map returns true and false after seeing the condition

```python
scores = [85, 92, 78, 65, 45]

ans = list(map(lambda x: x>80,scores))
print(ans)
```

# Reduce
reduce(function, iterable)

```python
from functools import reduce

scores = [1,2,3,4]

sum_scores = reduce(lambda x,y: x+y,scores)
print(sum_scores)
```

# Problems with Text mode

```python

# binary files like image not work
with open('Books.png', 'r') as f:
  f.read()
```

```python
# write binaray mode helps us to work with image
with open('Books.png', 'rb') as f:
  f.read()
  with open('books_copy.png', 'wb') as f1:
      f1.write(f.read())    # copying image 1 to image 2
```

# Serialization and Deserialization
- Serialization - process of converting python data types to JSON format
- Deserialization - process of converting JSON to python data types

```python
# to work with dictionary, we use  Serialization and Deserialization
```

```python
a = {
    'name': 'preta',
    'age': 22
}


# writing in sample7 as json format -> serialization
import json
with open('sample7.txt','w') as f:
    json.dump(a,f)
```

```python
# deserialization
with open('sample7.txt','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
    
```

so, in here we can access it as dict

# A program can have errors at two different times:
- Compile time → Syntax errors (while writing the code, will show red mark)
- Run time or Execution time → Exceptions

# Syntax Error
- Code is written incorrectly (breaks language rules/grammar)
- Caught by the compiler or interpreter
- Fixed by correcting the code structure -> By Debuging

# Examples of Syntax Error
- Missing symbols (colon, brackets, etc.)
- Misspelled keywords
- Wrong indentation
- Empty blocks (if/else, loops, classes, functions)

# Exceptions
Errors that happen while the program is running.
- Python raises them automatically.
- Handle them to keep your program from crashing.<br>
Examples
- Memory overflow
- Dividing a number by 0 => Logical error
- Database connection errors<br>
Why we need Exception Handling
- User Experience
- Security reason

```python

try:
  with open('sample111.txt', 'r') as f:
    print(f.read())
except:
  print("File not found")

print('DOing something...')
```

```python
# to see what type of error is in there:

try:
  with open('sample1111.txt', 'r') as f:
    print(f.read())
    print(1/0)
except Exception as e:
  print(e.with_traceback)
```

```python

# else
try:
  f = open('sample.txt')
except Exception as e:
  print(e)
else:
  # sure here no exception error can occure
  print(f.read())
```

try–except–else structure-এ else ব্লকটি তখনই execute হয় যখন try ব্লকে কোনো exception হয় না।
<br>flow:<br>
- try → success → else → run<br>
- try → exception → except → run

```python

# finally
try:
  f = open('sample2222.txt')
except Exception as e:
  print(e)
else:
  # sure here no exception error can occure
  print(f.read())
finally:
  print('Always executed')
  
  
# error hok ba na houk, finally te jabei
```

# raise Exception
- Python raises exceptions automatically when something goes wrong.
- You can also manually raise an exception using raise.
- You can pass a message to explain why the exception happened.

```python
raise NameError("Amar icca")
```

```python

# --- DEMO: raising custom errors ---
def require_positive(n):
    if n <= 0:
        raise ValueError("n must be > 0")
    return n

def dividedbyn(n):
  if n==0:
    raise ZeroDivisionError("0 diye vag kora jabe nah")
  return 100/n;

try:
    # print(require_positive(-1))
    print(dividedbyn(0))
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print("Caught:", e)
```
