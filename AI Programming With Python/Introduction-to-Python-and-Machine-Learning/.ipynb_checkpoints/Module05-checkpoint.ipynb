# Set

```python
# mutable, unique elements
# unordered. no indexing and slicing
```

```python
A = {1,2,3,4}
print(type(A))
```

```python
B = []
print(type(B))

C = {}
print(type(C))  # empty set evave declear kora jayna
```

```python
# so to declear empty set:
D = set()
print(type(D))
```

```python
# no duplicates in set

S = { 1,2,4,12,2,1,2}

print(S)
```

```python
# adding value as mutable
S = { 1,2,4,12,2,1,2}

S.add(123)
print(S)

S.pop()  # removes 1st value of 0th position
print(S)

S.remove(4)
print(S)
```

```python
# union and intersection kora jay

a = {1,2,3,4,5}
b = {1,2,7,8,9}

print(a.union(b))
print(b.intersection(a))
```

```python
# disjoint means no common value in two sets. that means no intersaction

print(a.isdisjoint(b))
```

```python
# subset means b  set er sob value a set e thakle, b set er subset holo a set

print(a.issubset(b))
```

# Dictionary

```python
# (key, value)

# key will be immutable data type.ex: int,string,float,tuple  # always should be immutable. that's why list is not used as key
# value will be anything. also dictionary
```

```python
dic = {}

print(type(dic))
```

```python
dic = {"name": "Phitron", "age": 20, "address": "Dhaka"}
print(type(dic),dic)
```

```python
# how too access? 2 way

print(dic["name"])
print(dic.get("name"))
```

```python
dic["name"] = "Phitron AI/ML"    # changeable

print(dic)
```

```python
dic = {"name": "Phitron", "age": 20,"age": 30, "address": "Dhaka"}  # same name e key 2 bar rakhle latest ta store hoy

print(dic)
```

```python
# insert value

dic["marks"] = 30

print(dic)


dic.update({"eng marks": 50})

print(dic)
```

```python
del dic["eng marks"]   # to delete
print(dic)
```

```python

dic_2 = dic.copy()
print(dic_2)
```

# Dictionary View Method

```python
dic = {"name": "Phitron", "age": 20,"age": 30, "address": "Dhaka"}


keys = dic.keys()
print(keys)

dic["math_num"] = 30

print(keys)  # though we haven't updated the keys value but it is auto updated
```

```python
# so, keys get real time updatation after 1st time defining
```

```python
values = dic.values()
print(values)  #etao same keys er moto
```

```python
# item key value pair gulo tuple akare chole ashe

items = dic.items()
print(items)
```

```python
# items e iterates o kora jay
```

# Dictionary Comprehension

```python
dic = {x:x**2 for x in range(11) if x%2==0}
print(dic)
```

```python
Co_ordinates = [(10,10.5),(20.5,192),(23,234)]
locations = ["dhaka","chittagong","mymenshingh"]

exact_location_dic ={co_ord:loc for co_ord,loc in zip(Co_ordinates,locations)}

print(exact_location_dic)
```

```python
#  Given a string , print the frequency of the words
string = """data science machine learning data analysis machine 
learning statistics data models data training data validation features
features labels preprocessing data augmentation models data optimization 
gradient descent neural networks data tensors matrices visualization 
exploration pandas numpy matplotlib seaborn scikitlearn tensorflow pytorch 
deployment inference production monitoring reproducibility experiments results 
metrics accuracy precision recall f1 cross validation data machine
"""

words = string.split()
count_dict = {}

for word in words:
    count_dict[word] = count_dict.get(word,0) + 1

for key, value in count_dict.items():
    print(f"{key} is present {value} times ")


```
