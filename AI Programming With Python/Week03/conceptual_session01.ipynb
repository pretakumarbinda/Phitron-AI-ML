Array Creation and Manipulation
- Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.
- Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0. [ Home work]

```python
import numpy as np
a = np.random.randint(1,21,(5,5))
a[:,2] = 1
print(a)
```

```python
b = np.random.randint(1,17,(4,4))
np.fill_diagonal(b,0)
print(b)
```

Array Indexing and Slicing
- Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
- Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border. (HW)

```python
c = np.random.randint(1,36, (6,6))

d = c[2:5, 1:4]
print(d)
```

```python
arr = np.random.randint(0, 100, size=(5, 5))

a= arr[:,0]
b = arr[:, -1]
c = arr[0,:]
d = arr[-1,:]
border_elements = np.concatenate((a, b, c, d))
print(border_elements)
```
