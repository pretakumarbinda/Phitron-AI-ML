# Mini Project

```python

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 67, 90, 45, 88]
}

df = pd.DataFrame(data)
df
```

```python
# adding grade column
def assign_grade(marks):
    if marks>=80:
        return "A"
    elif marks>=70:
        return "B"
    elif marks>=60:
        return "C"
    elif marks>=50:
        return "D"
    else:
        return "F"
    
df["Grade"] = df["Marks"].apply(assign_grade)
```

```python
df
```

```python
# filtering
top_st = df[df["Marks"]>= 80]
top_st
```

```python
# sorting
sorted_df = df.sort_values(by= "Marks", ascending= False)
sorted_df
```

```python
avg_marks = df["Marks"].mean()
```
