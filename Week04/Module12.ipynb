# filtering on string

````python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
````

````python
df
````

````python
df.loc[df["City"].str.contains("New",case=False)]
````

```python

the r before the string (r"^Los") means “raw string” in Python.

🔹 What is a raw string?

A raw string tells Python:

“Treat everything inside this string exactly as it is — don’t interpret special escape characters.”

````python
# starts with Los

df.loc[df["City"].str.contains(r"^Los")]

````

````python
# ends with rk

df.loc[df["City"].str.contains(r"rk$")]
````

````python
# name starts with a vouel

df.loc[df["Name"].str.contains(r"^[AEIOUaeiou]")]
````

````python
# cities contains "Los" or "New"

df.loc[df["City"].str.contains(r"Los|New")]
````

# Handling duplicate values

````python
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David", "Bob"],
    "City": ["New York", "London", "Paris", "New York", "Tokyo", "London"],
    "Score": [85, 90, 78, 70, 95, 90]
}

df = pd.DataFrame(data)
df
````

````python
df.duplicated()
````

````python
df.duplicated().sum()    # number of duplicate rows
````

````python
df.drop_duplicates(inplace=True)
df
````

````python
# to delete column wise duplicates

# based on name

df.drop_duplicates(subset=["Name"], inplace=True)
df
````

````python
df.drop_duplicates(subset=["Name"], keep="last", inplace=True) # to keep the last occurrence of the duplicate value in the "Name" column and delete the rest
````

# Apply Function on DF

````python
df
````

````python
# min max scaling

min_val = df['total_marks'].min()
max_val = df['total_marks'].max()
df['total_marks_scaled'] = df['total_marks'].apply(lambda x: (x - min_val)/(max_val - min_val))
````

````python
df
````

````python
def grading_system(x):
    if x >= 250:
        return 'A'
    elif x >= 200:
        return 'B'
    elif x >= 150:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['total_marks'].apply(grading_system)
````

````python
df
````

````python
def marking_system(df):
    a = df['Data Structure Marks'] *2
    b = df['Algorithm Marks'] * 3
    c = df['Python Marks'] * 4
    return a+b+c

df['total_marks_weighted'] = df.apply(marking_system, axis=1)
````

# groupby in pandas

````python
df = pd.read_csv("student_completed_data.csv")
````

````python
df.info()
````

````python
import numpy as np

df.loc[df['FullName'] == 'Afsana Mimi', 'Total Marks'] = np.nan
df.loc[df['FullName'] == 'Sadia Chowdhury', 'Total Marks'] = np.nan
df.loc[df['FullName'] == 'Alif Rahman', 'Total Marks'] = np.nan
````

````python
instructor_group = df.groupby("Instructor")
instructor_group
````

````python
for name, group in instructor_group:
    print(name)
    display(group)
````

````python
instructor_group.min()
````

````python
instructor_group.describe()
````

````python
instructor_group.get_group("Mr. David")
````

````python
mean_marks = instructor_group['Total Marks'].mean()
````

````python
df['Total Marks']=df["Total Marks"].fillna(df["Instructor"].map(mean_marks))
````
