```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_students = pd.read_csv("students.csv")
df_sales = pd.read_csv("sales.csv")
```

```python
df_sales.head()
```

```python
#1
sns.lineplot(data=df_sales, x="date", y="sales")

```

```python
#2

sns.scatterplot(data=df_students, x="hours_studied", y="marks", hue="gender")
```

```python
#3  Create a grid of scatterplots (hours_studied vs marks) faceted by subject.

sns.relplot(data = df_students, x = "hours_studied", y="marks", col="subject", kind="scatter")
```

```python
# 4 : Plot a histogram of marks with a KDE overlay.
sns.histplot(data=df_students, x="marks", kde=True )
```

```python
#5 Show the KDE curve for hours_studied.

sns.kdeplot(data=df_students, x="hours_studied", fill=True)
```

```python
# 6 Display the count of records per subject.

sns.countplot(data=df_students, x="subject")
```

```python
sns.histplot(data=df_students, x="subject")
```

```python
# 7 Draw a regression line for hours_studied and marks.

sns.regplot(data=df_students, x="hours_studied", y="marks")
```

```python
# 8 Create a pairplot for students' dataset. 

sns.pairplot(df_students, hue="subject")
```
